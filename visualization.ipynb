```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
data = pd.read_csv("narendra_modi.csv")

# Calculate total sentiment scores
total_pos = data['pos'].sum()
total_neu = data['neu'].sum()
total_neg = data['neg'].sum()

# Bar positions
x = np.arange(3)

# Bar width
width = 0.2

# Plotting
fig, ax = plt.subplots(figsize=(8, 6))

# Positive Sentiment
ax.bar(x - width, [total_pos, total_neu, total_neg], width, label='Sentiment Scores', color=['green', 'gray', 'red'])

# Setting labels and title
ax.set_xlabel('Sentiment')
ax.set_ylabel('Total Score')
ax.set_title('Total Sentiment Scores for Narendra Modi')
ax.set_xticks(x)
ax.set_xticklabels(['Positive', 'Neutral', 'Negative'])
ax.legend()

plt.tight_layout()
plt.savefig('nmodi1.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV data into a DataFrame
df = pd.read_csv('narendra_modi.csv')

# Calculate the sentiment polarity score
df['sentiment_polarity'] = df['pos'] - df['neg']

# Detect outliers using Interquartile Range (IQR) method
Q1 = df['sentiment_polarity'].quantile(0.25)
Q3 = df['sentiment_polarity'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

# Identify outliers
outliers = df[(df['sentiment_polarity'] < lower_bound) | (df['sentiment_polarity'] > upper_bound)]
print(f"Number of outliers detected: {len(outliers)}")

# Print the outliers
print("\nOutliers:")
print(outliers[['sentiment_polarity', 'pos', 'neg', 'likes', 'views']])

# Box plot to visualize outliers
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='sentiment_polarity')
plt.title('Sentiment Polarity Outliers')
plt.savefig('nmodi4.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV data into a DataFrame
df = pd.read_csv('kejriwal_positive_punjabi.csv')

# Calculate the sentiment polarity score
df['sentiment_polarity'] = df['pos'] - df['neg']

# Detect outliers using Interquartile Range (IQR) method
Q1 = df['sentiment_polarity'].quantile(0.25)
Q3 = df['sentiment_polarity'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

# Identify outliers
outliers = df[(df['sentiment_polarity'] < lower_bound) | (df['sentiment_polarity'] > upper_bound)]
print(f"Number of outliers detected: {len(outliers)}")

# Print the outliers
print("\nOutliers:")
print(outliers[['sentiment_polarity', 'pos', 'neg', 'likes', 'views']])

# Box plot to visualize outliers
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='sentiment_polarity')
plt.title('Sentiment Polarity Outliers for Arvind Punjab')
plt.savefig('parvind2.png')
plt.show()
```
